```python
# Install required packages (run if needed)
# !pip install requests pandas matplotlib seaborn

import requests
import json
import pandas as pd
from IPython.display import Markdown, display
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")

print("ğŸ“¦ Packages loaded successfully!")
```

```python
# API Configuration
API_BASE_URL = "https://local-market-intelligence-api-1040656024374.us-central1.run.app"

def analyze_market(business_type, location, context=""):
    """Call the Local Market Intelligence API"""
    url = f"{API_BASE_URL}/analyze"
    payload = {
        "business_type": business_type,
        "location": location,
        "additional_context": context
    }
    response = requests.post(url, json=payload)
    return response.json()

# Test API connectivity
try:
    health_check = requests.get(f"{API_BASE_URL}/health")
    print(f"âœ… API Status: {health_check.json()}")
except Exception as e:
    print(f"âŒ API Connection Error: {e}")
```

## ğŸµ Coffee Shop Market Analysis - San Francisco

**Use Case**: Market entry analysis for premium coffee concept

```python
# Analyze coffee shop market in San Francisco
print("ğŸ” Analyzing coffee shop market in San Francisco...")
print("â³ This may take 10-30 seconds...")

sf_coffee = analyze_market(
    "coffee shops", 
    "San Francisco, CA",
    "Focus on premium market opportunities and geographic expansion"
)

# Display the comprehensive report
if sf_coffee.get('status') == 'success':
    print("âœ… Analysis Complete!")
    display(Markdown(sf_coffee['analysis']['comprehensive_report']))
else:
    print(f"âŒ Error: {sf_coffee.get('message', 'Unknown error')}")
```

## ğŸ½ï¸ Restaurant Market Analysis - Austin, TX

**Use Case**: Investment due diligence for restaurant acquisition

```python
# Analyze restaurant market in Austin
print("ğŸ” Analyzing restaurant market in Austin, TX...")
print("â³ This may take 10-30 seconds...")

austin_restaurants = analyze_market(
    "restaurants", 
    "Austin, TX",
    "Investment feasibility and competitive positioning analysis"
)

if austin_restaurants.get('status') == 'success':
    print("âœ… Analysis Complete!")
    display(Markdown(austin_restaurants['analysis']['comprehensive_report']))
else:
    print(f"âŒ Error: {austin_restaurants.get('message', 'Unknown error')}")
```

```python
# Extract and compare key metrics
if sf_coffee.get('status') == 'success' and austin_restaurants.get('status') == 'success':
    sf_metrics = sf_coffee['analysis']['summary_metrics']
    austin_metrics = austin_restaurants['analysis']['summary_metrics']

    comparison_data = {
        'Market': ['San Francisco Coffee', 'Austin Restaurants'],
        'Total Businesses': [sf_metrics['total_businesses_analyzed'], austin_metrics['total_businesses_analyzed']],
        'Opportunities': [sf_metrics['opportunities_identified'], austin_metrics['opportunities_identified']],
        'Geographic Clusters': [sf_metrics['geographic_clusters'], austin_metrics['geographic_clusters']]
    }

    comparison_df = pd.DataFrame(comparison_data)
    display(comparison_df)
    
    # Simple visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Businesses vs Opportunities
    ax1.bar(comparison_df['Market'], comparison_df['Total Businesses'], alpha=0.7, label='Total Businesses')
    ax1.set_title('Total Businesses Analyzed')
    ax1.set_ylabel('Count')
    plt.setp(ax1.get_xticklabels(), rotation=45)
    
    # Opportunities
    ax2.bar(comparison_df['Market'], comparison_df['Opportunities'], alpha=0.7, color='orange', label='Opportunities')
    ax2.set_title('Opportunities Identified')
    ax2.set_ylabel('Count')
    plt.setp(ax2.get_xticklabels(), rotation=45)
    
    plt.tight_layout()
    plt.show()
else:
    print("Cannot compare - one or both analyses failed")
```

## ğŸ“Š Multi-Market Expansion Analysis

**Use Case**: Chain expansion planning across multiple cities

```python
# Analyze multiple markets for expansion planning
markets = [
    ("coffee shops", "Seattle, WA"),
    ("coffee shops", "Portland, OR"), 
    ("coffee shops", "Denver, CO"),
    ("coffee shops", "Austin, TX")
]

expansion_results = []
print("ğŸš€ MULTI-MARKET EXPANSION ANALYSIS")
print("=" * 50)

for business_type, location in markets:
    print(f"ğŸ” Analyzing {location}...")
    try:
        result = analyze_market(business_type, location, "Expansion opportunity assessment")
        
        if result.get('status') == 'success':
            metrics = result['analysis']['summary_metrics']
            expansion_results.append({
                'City': location.split(', ')[0],
                'Total_Competitors': metrics['total_businesses_analyzed'],
                'Opportunities': metrics['opportunities_identified'],
                'Geographic_Clusters': metrics['geographic_clusters']
            })
            print(f"  âœ… {metrics['total_businesses_analyzed']} competitors, {metrics['opportunities_identified']} opportunities")
        else:
            print(f"  âŒ Analysis failed for {location}")
            
    except Exception as e:
        print(f"  âŒ Error analyzing {location}: {str(e)}")

# Create expansion opportunity matrix
if expansion_results:
    expansion_df = pd.DataFrame(expansion_results)
    print(f"\nğŸ“Š EXPANSION ANALYSIS RESULTS:")
    display(expansion_df)
else:
    print("No successful analyses to display")
```

```python
# Visualize expansion opportunities
if expansion_results and len(expansion_results) > 0:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    # Competition vs Opportunities scatter plot
    ax1.scatter(expansion_df['Total_Competitors'], expansion_df['Opportunities'], 
               s=expansion_df['Geographic_Clusters']*20, alpha=0.7, c='skyblue', edgecolors='navy')
    ax1.set_xlabel('Total Competitors')
    ax1.set_ylabel('Opportunities Identified')
    ax1.set_title('Market Competition vs Opportunities\n(Bubble size = Geographic Clusters)')
    ax1.grid(True, alpha=0.3)
    
    for i, city in enumerate(expansion_df['City']):
        ax1.annotate(city, (expansion_df['Total_Competitors'].iloc[i], expansion_df['Opportunities'].iloc[i]),
                    xytext=(5, 5), textcoords='offset points', fontsize=9)

    # Opportunity score by city
    bars = ax2.bar(expansion_df['City'], expansion_df['Opportunities'], color='lightcoral', alpha=0.7)
    ax2.set_title('Opportunities by City')
    ax2.set_ylabel('Number of Opportunities')
    ax2.grid(True, alpha=0.3, axis='y')
    plt.setp(ax2.get_xticklabels(), rotation=45)
    
    # Add value labels on bars
    for bar in bars:
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'{int(height)}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()
    
    # Best opportunity recommendation
    best_city = expansion_df.loc[expansion_df['Opportunities'].idxmax()]
    print(f"\nğŸ† BEST EXPANSION OPPORTUNITY: {best_city['City']}")
    print(f"   â€¢ {best_city['Opportunities']} opportunities identified")
    print(f"   â€¢ {best_city['Total_Competitors']} competitors")
    print(f"   â€¢ {best_city['Geographic_Clusters']} geographic clusters")
else:
    print("No data available for visualization")
```

## ğŸ¯ Interactive Analysis

Run the cell below for an interactive market analysis experience!

```python
def interactive_market_analysis():
    """Menu-based interactive analysis - reliable in all Jupyter environments"""
    
    business_types = {
        '1': 'coffee shops',
        '2': 'restaurants', 
        '3': 'retail stores',
        '4': 'fitness centers',
        '5': 'salons',
        '6': 'bakeries'
    }
    
    cities = {
        '1': 'San Francisco, CA',
        '2': 'New York, NY',
        '3': 'Austin, TX', 
        '4': 'Seattle, WA',
        '5': 'Miami, FL',
        '6': 'Denver, CO'
    }
    
    print("ğŸ¢ LOCAL MARKET INTELLIGENCE API - Interactive Analysis")
    print("=" * 60)
    
    print("\nğŸ“‹ Select Business Type:")
    for key, value in business_types.items():
        print(f"  {key}. {value.title()}")
    
    biz_choice = input("\nEnter choice (1-6): ").strip()
    business_type = business_types.get(biz_choice, 'coffee shops')
    
    print(f"\nğŸ“ Select Location:")
    for key, value in cities.items():
        print(f"  {key}. {value}")
    
    city_choice = input("\nEnter choice (1-6): ").strip()
    location = cities.get(city_choice, 'San Francisco, CA')
    
    context = input("\nğŸ’¬ Enter analysis context (or press Enter for default): ").strip()
    if not context:
        context = "Market analysis and opportunity assessment"
    
    print(f"\nğŸ” Analyzing {business_type} in {location}...")
    print("â³ This may take 10-30 seconds...")
    
    try:
        result = analyze_market(business_type, location, context)
        
        if result.get('status') == 'success':
            # Show key metrics
            metrics = result['analysis']['summary_metrics']
            print(f"\nâœ… Analysis Complete!")
            print("=" * 50)
            print(f"ğŸ“Š QUICK METRICS:")
            print(f"â€¢ Total Competitors: {metrics['total_businesses_analyzed']}")
            print(f"â€¢ Opportunities Identified: {metrics['opportunities_identified']}")
            print(f"â€¢ Geographic Clusters: {metrics['geographic_clusters']}")
            print(f"â€¢ Data Sources: {', '.join(metrics['data_sources'])}")
            
            # Show market dashboard
            print(f"\nğŸ“ˆ MARKET DASHBOARD:")
            print(result['analysis']['market_dashboard'])
            
            # Option to see full report
            show_full = input(f"\nğŸ“‹ Show full comprehensive report? (y/n): ").strip().lower()
            if show_full in ['y', 'yes']:
                print("\n" + "="*60)
                print("ğŸ“‹ COMPREHENSIVE BUSINESS INTELLIGENCE REPORT")
                print("="*60)
                display(Markdown(result['analysis']['comprehensive_report']))
                return result
            else:
                print(f"\nâœ… Analysis complete! Summary shown above.")
                return None
        else:
            print(f"âŒ Error: {result.get('message', 'Unknown error')}")
            return None
        
    except Exception as e:
        print(f"âŒ Error: {str(e)}")
        return None

# Run the interactive analysis
interactive_market_analysis();
```

## ğŸš€ Quick Analysis Functions

```python
def quick_analysis(business_type, location):
    """Quick one-line analysis for testing"""
    print(f"ğŸ” Quick analysis: {business_type} in {location}")
    try:
        result = analyze_market(business_type, location)
        
        if result.get('status') == 'success':
            metrics = result['analysis']['summary_metrics']
            
            competition_level = (
                "HIGH" if metrics['total_businesses_analyzed'] > 50 
                else "MEDIUM" if metrics['total_businesses_analyzed'] > 20 
                else "LOW"
            )
            
            print(f"ğŸ“Š {metrics['total_businesses_analyzed']} competitors | {metrics['opportunities_identified']} opportunities | Competition: {competition_level}")
        else:
            print(f"âŒ Analysis failed: {result.get('message', 'Unknown error')}")
    except Exception as e:
        print(f"âŒ Error: {str(e)}")
    
    return None

# Quick test examples
print("ğŸ¯ QUICK ANALYSIS EXAMPLES:")
print("-" * 40)
quick_analysis("coffee shops", "Portland, OR")
quick_analysis("restaurants", "Austin, TX") 
quick_analysis("fitness centers", "Miami, FL")
```
